# Grficos con Bokeh

```python
from bokeh.plotting import figure, output_file, save
output_file('data_science_popularity.html')
```

```python
p = figure(title='data science', x_axis_label='Mes', y_axis_label='data science')
p.line(df['Mes'], df['data science'], legend='popularity', line_width=2)
save(p)
```

```python
from bokeh.plotting import figure, gridplot, output_file, save
import pandas as pd

# Se asume que el DataFrame 'df' ya est definido

output_file('multiple_graphs.html')

s1 = figure(width=250, height=250, title='data science')
s1.circle(df['Mes'], df['data science'], size=10, color='navy', alpha=0.5)

s2 = figure(width=250, height=250, x_range=s1.x_range, y_range=s1.y_range, title='machine learning')
s2.triangle(df['Mes'], df['machine learning'], size=10, color='red', alpha=0.5)

s3 = figure(width=250, height=250, x_range=s1.x_range, title='deep learning')
s3.square(df['Mes'], df['deep learning'], size=5, color='green', alpha=0.5)

p = gridplot([[s1, s2, s3]])
save(p)
```

```python
import geopandas as gpd
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my") # Especifica tu propio valor de user_agent

df2 = pd.read_csv('mapa.csv')
df2.dropna(axis=0, inplace=True)

geocoded = df2['Pas'].apply(lambda x: geolocator.geocode(x)) # Geocodifica cada pas

df2['geometry'] = geocoded.apply(lambda location: location.point if location else None)
df2['Latitude'] = df2['geometry'].apply(lambda point: point.latitude if point else None)
df2['Longitude'] = df2['geometry'].apply(lambda point: point.longitude if point else None)
```

```python
m3 = folium.Map(location=[39.326234,-4.838065], tiles='openstreetmap', zoom_start=3)

def color_producer(val):
	if val <= 50:
		return 'red'
	else:
		return 'green'

for i in range(0,len(df2)):
	folium.Circle(location=[df2.iloc[i]['Latitude'], df2.iloc[i]['Longitude']], radius=5000*df2.iloc[i]['data science'], color=color_producer(df2.iloc[i]['data science'])).add_to(m3)

m3.save('map3.html')
```
