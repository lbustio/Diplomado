# Representación de la información mediante gráficos

## Importar librerías para el uso

```python
import numpy as np # libreria para el trabajo con matrices
import matplotlib.pyplot as plt # libreria para la visualizacion
import random # libreria par el manejo de numeros aleatorios
```

## Generar datos sintéticos aleatorios

```python
rng = 10
x = range(0, rng)
y = list() # lluvia en el 2019
z = list() # lluvia en el 2020
```

```python
for i in range(rng):
  y.append(random.randint(0, 9))
  z.append(random.randint(0,15))
```

## Gráfico de área

```python
# Encabezados y personalizacion general del grafico
plt.title("Precipitaciones por año")

# Dibujos de los graficos
plt.fill_between(x, z)
plt.fill_between(x, y)

# Personalizaciones finas
# Personalizando X
plt.xlabel("x label")
plt.tick_params(axis='x', colors='red', direction='out', length=7, width=2)
plt.xticks(x, ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sept", "Oct"])

# Personalizando Y
plt.ylabel("y label")

# Mostrar el resultado
plt.show()
```

## Gráfico de barras

```python
# Generar datos
height = [3, 12, 5, 18, 45]
bars = ["Clase A", "Clase B", "Clase C", "Clase D", "Clase E"]
y_pos = np.arange(len(bars))

# Crear el gráfico de barras
plt.bar(y_pos, height)

# personalizacion
plt.xticks(y_pos, bars)

# mostrar gráfico
plt.show()
```

## Boxplot

```python
# instalar librería gráfica seaborn
#!pip install seaborn
import seaborn as sns

# importar datos
iris_df = sns.load_dataset("iris")

# representar en boxplot a iris
sns.boxplot(x = iris_df["species"], y = iris_df["sepal_length"])
sns.boxplot(x = iris_df["species"], y = iris_df["sepal_width"])
plt.show()
```

## Radial graph

```python
import plotly.graph_objects as go

categories = ['processing cost','mechanical properties','chemical stability',
              'thermal stability', 'device integration']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[1, 5, 2, 2, 3],
      theta=categories,
      fill='toself',
      name='Product A'
))

fig.add_trace(go.Scatterpolar(
      r=[4, 3, 2.5, 1, 2],
      theta=categories,
      fill='toself',
      name='Product B'
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 5]
    )),
  showlegend=False
)

fig.show()
```
